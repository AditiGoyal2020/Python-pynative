# Matplotlib

```python
import pandas as pd
import matplotlib.pyplot as plt  
```

```python
#Read Total profit of all months and show it using a line plot
#Total profit data provided for each month. Generated line plot must include the following properties: –
#X label name = Month Number
#Y label name = Total profit

Data=pd.read_csv("company_sales_data.csv")
monthlist= Data["month_number"]
profitlist=Data["total_profit"]
plt.plot(monthlist,profitlist)
```

```python
plt.plot(Data.month_number,Data.total_profit,'green')
```

```python
# Get Total profit of all months and show line plot with the following Style properties
plt.plot(Data.month_number,Data.total_profit,'g--')
```

```python
# Read all product sales data and show it  using a multiline plot
plt.plot(Data.month_number,Data.facecream,'g-*')
plt.plot(Data.month_number,Data.facewash,'b*-')
plt.plot(Data.month_number,Data.toothpaste,'b*-')
plt.plot(Data.month_number,Data.bathingsoap,'r*-')
plt.plot(Data.month_number,Data.shampoo,'y*-')
plt.plot(Data.month_number,Data.moisturizer,'b*-')
```

```python
# Read toothpaste sales data of each month and show it using a scatter plot

plt.scatter(Data.month_number,Data.toothpaste, label='Tooth paste sales data')
plt.xlabel=('month_number')
plt.ylabel=('toothpaste')
plt.title ('Tooth paste sales data')
plt.legend(loc='upper left')
plt.grid(True, linewidth=1, linestyle="--")
```

```python
# Read face cream and facewash product sales data and show it using the bar chart
plt.bar([a-0.25 for a in Data.month_number],Data.facecream,width=0.25,label='Face Cream sales data',align='edge')
plt.bar([a+0.25 for a in Data.month_number], Data.facewash, width=0.25, label ='Face Wash sales data', align='edge')
plt.legend(loc='upper left')
plt.title("sales data")
plt.xlabel('Month_number')
plt.ylabel('Sales units in number')

```

```python
# Read sales data of bathing soap of all months and show it using a bar chart. Save this plot to your hard disk
plt.bar(Data.month_number,Data.bathingsoap)
plt.title(' Sales data')
plt.grid(False, linewidth= 1, linestyle="--")
plt.title('bathingsoap sales data')
plt.savefig('D:\Imparticus\DSP -Online\PYTHON\Data Files',dpi=150)
```

```python
#Read the total profit of each month and show it using the histogram to see most common profit ranges
profit_range =[150000, 175000,200000,225000,250000,300000,325000,350000,375000,400000,425000]
plt.hist(Data.total_profit, profit_range,label = 'Profit data')
plt.legend(loc='upper left')
plt.title('Profit data')
```

```python
# Calculate total sale data for last year for each product and show it using a Pie char
newdata =Data.iloc[:,1:7]
```

```python
yeardata=[]
for i in newdata:
    yeardata.append(sum(Data[i]))

newdata1=list(newdata)
```

```python
plt.pie(yeardata, labels=newdata1, autopct='%1.1f%%')
plt.title("Sales data distributution")
plt.legend(loc="lower right")
plt.show()
```

```python
# Read Bathing soap facewash of all months and display it using the Subplot
plt.subplot(2,1,1)
plt.plot(Data.month_number,Data.bathingsoap,linewidth=3,color='r')
plt.title("Sales data of  a Bathingsoap")
plt.subplot(2,1,2)
plt.plot(Data.month_number,Data.facewash,linewidth=3,color='b')
plt.title("Sales data of  a facewash")
```
